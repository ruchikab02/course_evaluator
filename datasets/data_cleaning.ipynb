```python
import pandas as pd
import numpy as np
gpa_data = pd.read_csv(r"datasets/fa2019-gpa.csv")
gpa_data.info()
```

```python
prereqs = pd.read_csv(r"datasets/uiuc-prerequisites.csv")
prereqs.info()
```

```python
to_drop = ['CRN', 'Course Title', 'Sched Type', 'A+', 'A', 'A-', 
          'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'W', 'Primary Instructor', 'Course ', 'Subject']
gpa_data["Course"] = gpa_data["Subject"].astype(str) + " " + gpa_data["Course "].astype(str)
gpa_data.drop(to_drop, inplace=True, axis=1)
print(gpa_data)
```

```python
print(gpa_data.shape)
print(prereqs.shape)
```

```python
group = gpa_data.groupby("Course")
df_mean = group.agg("mean").reset_index()
```

```python
df_mean
print(df_mean.shape)
```

```python
prereqs
```

```python
merged_data = pd.merge(df_mean, prereqs, how='left', on='Course')
```

```python
merged_data
```

```python
merged_data["Average Grade"] = 4.0 - merged_data["Average Grade"]
```
